# Upload my clean data and encode it

```python
import pandas as pd
from sentence_transformers import SentenceTransformer

# Load CSV
df = pd.read_csv("/content/cleaned_clothes_shop_chatbot.csv")

documents = df['patterns'].tolist()  # Use patterns as docs to embed

# Responses associated with each pattern (same order)
responses = df['responses'].tolist()

# Initialize embedding model
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Embed patterns
doc_embeddings = embedder.encode(documents, convert_to_numpy=True)
```

```python
import faiss
import numpy as np

d = doc_embeddings.shape[1]
index = faiss.IndexFlatIP(d)
faiss.normalize_L2(doc_embeddings)
index.add(doc_embeddings)

def retrieve_response(query, top_k=3):
    query_emb = embedder.encode([query], convert_to_numpy=True)
    faiss.normalize_L2(query_emb)
    distances, indices = index.search(query_emb, top_k)
    return [responses[i] for i in indices[0]]

def truncate_text(text, max_len=150):
    return text if len(text) <= max_len else text[:max_len] + "..."
```

```python
def generate_answer(query):
    retrieved = retrieve_response(query, top_k=3)
    retrieved = [truncate_text(r) for r in retrieved]
    context = "\n".join(retrieved)

    prompt = f"""
You are an e-commerce chatbot that helps customers with their questions about products, orders, shipping, returns, and store policies.

Use the following information from the knowledge base to answer the question as accurately and helpfully as possible:

Context:
{context}

Customer question: {query}

Answer politely and clearly in one paragraph without continuing the conversation:
"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        do_sample=False,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id,
    )
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer[len(prompt):].strip()
```
